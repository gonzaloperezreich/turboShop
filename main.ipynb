```python
import spacy
import numpy as np
# Load the trained model
output_dir = "./modelo_repuestos_ner"
nlp = spacy.load(output_dir)

# Initialize a list to store the extracted entities
savedEntities = []

# Process each test text
for text in repuestos:
    docUpper = nlp(text.upper())  # Process uppercase text
    docLower = nlp(text.lower())  # Process lowercase text

    # Collect entities from both versions
    entitiesUpper = {(ent.text, ent.label_) for ent in docUpper.ents}
    entitiesLower = {(ent.text.upper(), ent.label_.upper()) for ent in docLower.ents}

    # Combine entities using a set to avoid duplicates
    combinedEntities = entitiesUpper.union(entitiesLower)

    # Convert the set back to a list (optional, if you need it as a list)
    combinedEntitiesList = list(combinedEntities)

    print(f"\nText: {text}")
    print("Entities:", combinedEntitiesList)

    # Append the combined entities to the savedEntities list
    savedEntities.append((text,combinedEntitiesList))
```
